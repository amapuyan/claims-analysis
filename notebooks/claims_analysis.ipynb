```python
merged_hl = pd.merge(df_header, df_line, on='ProspectiveClaimId', how='left')

line_counts = (
    merged_hl.groupby('ProspectiveClaimId')
    .size()
    .reset_index(name='LineCount')
)

charges_by_claim = (
    merged_hl.groupby('ProspectiveClaimId')['Charges']
    .sum()
    .reset_index(name='TotalCharges')
)

claim_summary = (
    df_header[['ProspectiveClaimId', 'BillingProviderNPI', 'BillingProvFirstName', 'BillingProvLastName']]
    .merge(line_counts, on='ProspectiveClaimId')
    .merge(charges_by_claim, on='ProspectiveClaimId')
)

high_line_claims = claim_summary[claim_summary['LineCount'] >= 5]

display(high_line_claims)
```

```python
# Q7 Diagnosis-Procedure Combination
# Merge
hl = pd.merge(df_header, df_line, on='ProspectiveClaimId', how='left')

# Merge results with code
full_claims = pd.merge(hl, df_code, on='ProspectiveClaimId', how='left')

#  Filter CPT 99291
cpt_99291 = full_claims[full_claims['HCPCS'] == '99291']

# Count diagnosis codes associated with 99291
top_dx_99291 = (
    cpt_99291['CodeValue']
    .value_counts()
    .head(10)
)

print("Most Common Diagnosis Codes with CPT 99291:")
display(top_dx_99291)
```

```python
#Q8 Charges by Payer
# Merge header and line
merged_hl = pd.merge(df_header, df_line, on='ProspectiveClaimId', how='left')

# find total charges per claim
total_charges_per_claim = (
    merged_hl.groupby('ProspectiveClaimId')['Charges']
    .sum()
    .reset_index(name='TotalClaimCharges')
)

# add claim charges back to Header
header_with_charges = df_header.merge(
    total_charges_per_claim,
    on='ProspectiveClaimId',
    how='left'
)

# Group by payer
payer_summary = (
    header_with_charges.groupby('PrimaryPayerName')
    .agg(
        TotalCharges=('TotalClaimCharges', 'sum'),
        AvgChargesPerClaim=('TotalClaimCharges', 'mean'),
        NumberOfClaims=('ProspectiveClaimId', 'nunique')
    )
    .sort_values('TotalCharges', ascending=False)
    .head(10)
)

display(payer_summary)
```

```python
# Part 4: Creative Analysis

# Q9 Own Analyis (Which providers bill for the most complex cases (measured by highest average number of diagnosis codes per claim)

# count diagnosis codes per claim, and tie them back to providers
header_code = pd.merge(
    df_header[['ProspectiveClaimId', 'BillingProvFirstName', 'BillingProviderNPI']],
    df_code[['ProspectiveClaimId', 'CodeValue']],
    on='ProspectiveClaimId',
    how='left'
)

dx_counts = (
    header_code.groupby(['ProspectiveClaimId', 'BillingProvFirstName', 'BillingProviderNPI'])
    .size()
    .reset_index(name='NumDiagnosisCodes')
)

# Avg diagnosis codes per provider
provider_complexity = (
    dx_counts.groupby(['BillingProvFirstName', 'BillingProviderNPI'])
    .agg(AverageDxCodes=('NumDiagnosisCodes', 'mean'),
         TotalClaims=('ProspectiveClaimId', 'nunique'))
    .reset_index()
)

# Identify top providers
top_complex_providers = provider_complexity.sort_values(
    'AverageDxCodes', ascending=False
).head(10)

display(top_complex_providers)

plt.figure(figsize=(10,6))
plt.barh(
    top_complex_providers['BillingProvFirstName'],
    top_complex_providers['AverageDxCodes']
)
plt.xlabel("Average Number of Diagnosis Codes")
plt.title("Top Providers by Case Complexity (Diagnosis Codes per Claim)")
plt.gca().invert_yaxis()
plt.show()
```
